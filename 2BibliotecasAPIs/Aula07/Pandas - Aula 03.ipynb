```python
import pandas as pd

dadosNegocio = pd.DataFrame(
    {
        'Ganhos':[0.0, 1500.0, 1700.0, 1200.0, 1000.0, 1200.0, 0.0],
        'Gastos':[0.0, 350.0, 500.0, 200.0, 600.0, 500.0, 0.0]
    }, index=['01/08', '02/08', '03/08', '04/08', '05/08', '06/08', '07/08']
)

dadosNegocio
```

```python
# o método drop remove uma ou mais linhas ou colunas
# o parâmetro axis = 0 indica que é pra remoção ser sobre linhas (axis = 1, colunas)

semFds = dadosNegocio.drop(['01/08', '07/08'], axis=0) # note que o drop retorna um dataframe novo com as linhas removidas


semFds
```

```python
dadosNegocio # o original não foi alterado!
```

```python
dadosReindexado = dadosNegocio.reset_index() # o reset_index cria novos índices numéricos, e transforma os índices antigos em uma coluna nova
```

```python
dadosReindexado
```

```python
dadosReindexado.rename(columns={'index':'Datas'}, inplace=True) # o parâmetro inplace muda o comportamento: ele modifica o dataframe ORIGINAL

dadosReindexado
```

```python
dadosReindexado.drop([0, 6], axis=0, inplace=True)

dadosReindexado
```

```python
dadosReindexado.reset_index() # note que ele criou uma coluna para os índices novos...
```

```python
dadosReindexado.reset_index(drop=True, inplace=True) # o parâmetro drop joga fora os índices velhos

dadosReindexado
```

```python
dadosReindexado['Lucro Liquido'] = dadosReindexado['Ganhos'] - dadosReindexado['Gastos'] # subtração elemento a elemento

dadosReindexado
```

```python
dadosReindexado['Fluxo'] = dadosReindexado['Lucro Liquido'].cumsum() # soma CUMULATIVA da coluna

dadosReindexado
```

```python
medalhas = pd.DataFrame(
{
    'Ouro':[39, 38, 27, 22, 20],
    'Prata':[41, 32, 14, 21, 28],
    'Bronze':[33, 18, 17, 22, 23]    
}, index=['EUA', 'China', 'Japão', 'Reino Unido', 'Rússia'])

medalhas
```

```python
medalhas['Total'] = medalhas['Ouro'] + medalhas['Prata'] + medalhas['Bronze'] # soma elemento a elemento...

medalhas
```

```python
medalhasReordenadas = medalhas.sort_values(by='Total') # ordenação segundo uma coluna específica (no caso, Total)

medalhasReordenadas # oops, saiu em ordem CRESCENTE...
```

```python
medalhasReordenadas = medalhas.sort_values(by='Total', ascending = False) # ordem decrescente

medalhasReordenadas
```

```python
# Voltando a nossa ordenação...

medalhasReordenadas.sort_values(by=['Ouro', 'Prata', 'Bronze'], ascending = False)

# quando o by é uma lista de colunas, ele utiliza a primeira, e em seguida usa a segunda como desempate, e assim sucessivamente...
```

5) Crie um programa para mostrar qual é o valor mais frequente de uma series e, em seguida, substitua todos os demais valores dela por "Outro valor".

```python
serie = pd.Series([1, 2, 3, 4, 2, 5, 6, 2])

serie
```

```python
serie[ serie != 2 ] = 'Outro valor'

serie
```

```python
# A solução anterior não está nem um pouco genérica! Vamos melhorar:

serie = pd.Series([1, 2, 3, 4, 2, 5, 6, 2])

moda = serie.mode()

moda
```

```python
serie[ serie != moda[0] ] = 'Outro valor'

serie
```

```python
serie2 = pd.Series([1, 2, 3, 4, 2, 5, 4, 6, 2, 4])

moda2 = serie2.mode()

moda2
```

```python
serie2[ ~(serie2.isin(moda2)) ] = 'Outro valor'

serie2
```
