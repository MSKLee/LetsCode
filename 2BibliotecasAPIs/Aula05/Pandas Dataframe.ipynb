# Merge

```python
df_left = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K3"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
    }
)


df_right = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    }
)
```

```python
df_left
```

```python
df_right
```

```python
df_resultado = pd.merge(df_left, df_right, on='key')
df_resultado
```

### Cruzamento de DataFrame com dois críterios

```python
df_left = pd.DataFrame(
    {
        "key1": ["K0", "K0", "K1", "K2"],
        "key2": ["K0", "K1", "K0", "K1"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
    }
)


df_right = pd.DataFrame(
    {
        "key1": ["K0", "K1", "K1", "K2"],
        "key2": ["K0", "K0", "K0", "K0"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    }
)
```

```python
df_left
```

```python
df_right
```

```python
df_resultado = pd.merge(df_left, df_right, on=['key1', 'key2']) # Caso as chaves fossem iguais
df_resultado
```

```python
df_resultado = pd.merge(df_left, df_right, left_on=['key1', 'key2'], right_on=['key1', 'key2']) # Caso as chaves fossem diferentes
df_resultado
```

```python
df_resultado = pd.merge(df_left, df_right, left_on=['key1', 'key2'], right_on=['key1', 'key2'], how='outer')
df_resultado
```

Questão #3
Enunciado

Crie um objeto DataFrame que contenha os dados da tabela abaixo, os quais se referem ao peso, altura e idade de cinco pessoas:  
Peso (kg) 	Altura (cm) 	Idade  
72 	180 	31  
80 	170 	21  
60 	175 	16  
90 	174 	28  
100 	185 	23  

Com base no DataFrame criado, realize o que pedido nos itens abaixo:  

(a) Visualize apenas os dados da coluna que contém a altura das pessoas.  
(b) Obtenha a idade média das pessoas.   
(c) Obtenha os dados das pessoas que estão abaixo da altura média.  
(d) Crie um nova coluna no DataFrame que contenha o IMC (Índice de Massa Corporal) de cada pessoa.  
(e) Remova a coluna que contém a idade das pessoas.  
(f) Salve esses dados do DataFrame em um arquivo csv.  


```python
import pandas as pd
import numpy as np


dic_dados3 = {
'peso' : [72, 80, 60, 90, 100],
'altura' : [180, 170, 175, 174, 185],
'idade' :[31, 21, 16, 28, 23]
}
    
dados3 = pd.DataFrame(dic_dados3)

media_altura = np.mean(dados3['altura'])

array3 = dados3.index.values[dados3.altura.values < media_altura]


def imc(linha3):
    imc_pessoa = (linha3['peso'])/(((linha3['altura'])/100)** 2)
    return imc_pessoa

dados_imc = dados3['imc']=dados3.apply(imc, axis = 1)

dados3_novo = dados3.drop(labels=None, axis=0, index=None, columns=['idade'])

print('Essa é a coluna das alturas das pessoas:\n',dados3['altura'])
print('A idade média das pessoas é:', np.mean(dados3['idade']))
print('Esses são os dados das pessoas que estão abaixo da altura média:\n',dados3.loc[array3])
print('Esses são os IMCs das pessoas:\n',dados_imc)
print('Esses são os dados sem a coluna "idade":\n', dados3_novo)

dados3.to_csv('dados3.csv', index= None)
dados3_novo.to_csv('dados3_novo.csv', index= None)
```

Questão #5

Enunciado

Carregue, em um objeto DataFrame, o mesmo dataset que você salvou na Questão 3. Em seguida,
  
(a) Adicione uma nova coluna nesse DF chamada Classificação, que contenha a classificação de cada indivíduo, de acordo com o seu IMC. Para isso, consulte a tabela abaixo.  
IMC 	Classificação 	Obesidade (grau)  
Menor que 18,5 	Magreza 	0  
Entre 18,5 e 24,9 	Normal 	0  
Entre 25,0 e 29,9 	Sobrepeso 	I  
Entre 30,0 e 39,9 	Obesidade 	II  
Maior que 40,0 	Obesidade Grave 	III  

(b) Salve em DataFrame no formato xlsx.


```python
import pandas as pd


dic_dados5 = {
'peso' : [72, 80, 60, 90, 100],
'altura' : [180, 170, 175, 174, 185],
'imc': [22.222222, 27.681661, 19.591837, 29.726516, 29.218408]
    
}
    
dados5 = pd.DataFrame(dic_dados5,index = ['Fernando', 'Thiago', 'Fabiano', 'Rodrigo', 'Rafael'])

def Classif(linha5):
    if linha5['imc']< 18.5:
        return 'Magreza'
    elif linha5['imc']>18.5 and linha5['imc']<24.9:
        return 'Normal'
    elif linha5['imc']>25 and linha5['imc']<29.9:
        return 'Sobrepeso'
    elif linha5['imc']>30 and linha5['imc']<39.9:
        return 'Obesidade'
    else:
        return 'Obesidade Grave'

dados5['Classificação'] = dados5.apply(Classif, axis=1)

dados5.to_excel('dados5.xlsx', index = None)
```
