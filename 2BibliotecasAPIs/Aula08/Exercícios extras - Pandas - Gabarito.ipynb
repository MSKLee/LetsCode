#### Exercício 3

Após tratar os dados duplicados, verifique se há dados faltantes no seu DataFrame. Trate cada coluna com dados faltantes separadamente, de maneira que não haja mais esse problema.

Dica: para preencher os dados, utilize o .fillna()

````python
df_casas.info()
````

````python
df_casas['FACHADA'] = df_casas['FACHADA'].fillna(df_casas['FACHADA'].mean())
df_casas['RUA'] = df_casas['RUA'].fillna(df_casas['RUA'].mode()[0])
df_casas['ANO_CONSTRUCAO'] = df_casas['ANO_CONSTRUCAO'].fillna(df_casas['ANO_CONSTRUCAO'].median())
df_casas['QUARTOS'] = df_casas['QUARTOS'].fillna(df_casas['QUARTOS'].mean())

df_casas = df_casas.drop('COR', axis=1)

df_casas
````

````python
df_casas.info()
````

#### Exercício 4

Utilize o método .describe() para analisar as colunas do seu DataFrame. Alguma coluna possui dados inconsistentes? Trate esse problema da maneira que achar adequada e, depois, monte uma pivot_table em que as linhas são o número de banheiros, as colunas são o número de quartos e os valores são os preçoes de venda agrupados pela mediana.

Dica: faz sentido uma casa com -10 banheiros?

**DESAFIO**: monte um segundo pivot_table, mas com a contagem proporcional de apartamentos em vez da mediana

````python
df_casas.describe()
````

````python
df_casas['BANHEIROS'] = df_casas['BANHEIROS'].replace(-1, 0)

df_casas.describe()
````

````python
df_casas.pivot_table(index='BANHEIROS', columns='QUARTOS', values='VALOR_VENDA', aggfunc=np.median)
````

````python
# Desafio

df_count = df_casas.pivot_table(index='BANHEIROS', columns='QUARTOS', values='VALOR_VENDA', aggfunc=np.size)
total = df_casas.shape[0]

# Dividimos a quantidade pelo total, para obtermos a proporção. O np.round limita o número de casas decimais
df_count_proporcional = np.round(df_count/total, 2)

df_count_proporcional
````

Depois de realizar todos os tratamentos necessários, analise os dados para responder as perguntas abaixo com VERDADEIRO ou FALSO. O seu código deve justificar as respostas:

4. Vendas associadas ao domínio de email do Yahoo são mais caras (V ou F) **(DESAFIO!)**


1.: F

````python
df_casas.groupby('ANO_CONSTRUCAO')['VALOR_VENDA'].mean().plot()
````

2.: V

````python
df_casas['CAT_TAMANHO'] = pd.qcut(df_casas['AREA'], 10)
````

````python
df_casas.groupby('CAT_TAMANHO')['VALOR_VENDA'].mean().plot(figsize=(12,4))
````

3.: V

````python
# Nesse caso, as casas com 0 banheiros se destacaram com uma maior área. Isso pode indicar
# que o nosso prenchimento de valores, substituindo -1 por 0, pode não ter sido correto.
# Na prática, poderíamos usar a área total para estimar a quantidade de banheiros nos casos
# de inconsistência. Como para os outros valores que não sofreram inferência a regra é verdadeira,
# ou seja, a área cresce com o número de banheiros, considerarei a afirmação como V.

df_casas.groupby('BANHEIROS')['AREA'].median().plot(figsize=(12,4))
````
