```python
import seaborn as sns
sns.set_theme()

%matplotlib inline
```

## Problema

```python
df_pinguins = sns.load_dataset('penguins')
df_pinguins.head()
```

```python
columns = {
    'species': 'espécie',
    'island': 'ilha',
    'bill_length_mm': 'comprimento_bico_mm',
    'bill_depth_mm': 'profundidade_bico_mm',
    'flipper_length_mm': 'comprimento_nadadeira_mm',
    'body_mass_g': 'massa_corpo_g',
    'sex': 'sexo'
    
}

df_pinguins.rename(columns = columns, inplace=True)

df_pinguins.head()
```

### countplot

A recomendação é de que se utilizem, em especial, as variáveis categóricas

Nosso dataset é desbalanceado em relação ao sexo?

```python
sns.countplot(data=df_pinguins, x='sexo');
```

Nosso dataset é balanceado em relação a ilhas?

```python
sns.countplot(data=df_pinguins, x='ilha');
plt.title('Ilhas', loc='left');#right
```

Nas ilhas existem desbalanceamentos em relação ao sexo?

```python
sns.countplot(data=df_pinguins, x='ilha', hue='sexo');
```

Existe diferença entre as espécies nas ilhas?

```python
sns.countplot(data=df_pinguins, x='ilha', hue='espécie');
```

```python
sns.countplot(data=df_pinguins, x='espécie');
```

### barplot

Será que o comprimento da nadadeira muda por espécie?

```python
df_pinguins.columns
```

```python
import pandas as pd

df_pinguins.groupby(by=['espécie', 'sexo'])['comprimento_nadadeira_mm'].size()
```

```python
sns.barplot(x='comprimento_nadadeira_mm', y='espécie',hue='sexo', data=df_pinguins);
plt.legend(loc='right');
```

### boxplot

```python
sns.boxplot(x='massa_corpo_g', data=df_pinguins, showmeans=True);
#showmeans = True
```

```python
sns.boxplot(x='massa_corpo_g', y='espécie', data=df_pinguins, showmeans=True);
```
