```python
# Importação das bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Para verificar a TCL, aumente o tamanho da amostra e o total de amostras
tam = 20  #tamanho de cada da amostra
total_sample = 100 #total de amostras 
dados = []
media_dados = []
for _ in range(total_sample):
    media = 0
    for i in range(tam):
        temp = np.random.randint(1,7)
        #temp = np.random.uniform()
        #temp = np.random.binomial(tam, 0.9)
        #temp = np.random.exponential(scale = 2)
              
        dados.append(temp)
        media += temp
    #print(media/tam)
    media_dados.append(media/tam)


media_dados = pd.Series(media_dados)
plt.plot(dados)
plt.show()
sns.histplot(dados)
plt.show()
sns.histplot(media_dados, kde=True)
#media/tam
```
